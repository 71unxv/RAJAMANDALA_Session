```python
import numpy as np
import pandas as pd
import lasio
import welly
import os, glob
import json
from shutil import copyfile
```

```python
def getWellPath(initialdir):
    """ 
    Search all the available .Las digital files and return its path for further directory references
    """    

    wellfolder = os.listdir(initialdir)
    largest_all = []
    for wellfolder_temp in wellfolder:
        path =initialdir +"\\" + wellfolder_temp + "\\Log Digital"
        if os.path.exists(path):
            if (len(os.listdir(path)) !=0):
#                 if os.path.getsize(path)>0:
                largest_all.append(sorted( (os.path.getsize(s), s) for s in glob.glob(path + "\\*.LAS") )[-1][1])
    print("Total Well: " + str(len(wellfolder)))
    print("Total Available .LAS data : " + str(len(largest_all)))
    return largest_all
    
def getUniqueMnemonic(AllWellPath):
    """ 
    unique Mnemonic Header from all well logs and return the unique Mnemonic header
    """
    
    header_DB = pd.DataFrame(columns = ['Header', 'Description'])

    mnemonic_all = []
    descr_all = []
    for WellPath in AllWellPath:
        Well = lasio.read(WellPath)
        print(WellPath)
        for mnemonic in Well.well.keys():
            descr_all = descr_all + [eval("Well.well." + mnemonic + ".__getitem__('descr')")]
            mnemonic_all = mnemonic_all + [mnemonic]
    header_DB['Header'] = mnemonic_all
    header_DB['Description'] = descr_all
    header_DB = header_DB.drop_duplicates()
    return header_DB

def batchReadMnemonicData(WellPathList, InitialPath, Mnemonic_head, Mnemonic_out_column):
    """ 
    Batch Screening and return the value of Mnemonic data from all well logs
    """
    
    columns_name = (["WellPathName", "WellPathName"] + Mnemonic_out_column)
    
    DB_list = []
    
    DB_list_temp = []

    for i in range(len(WellPathList)):
        WellPath = WellPathList[i]
        Well = lasio.read(WellPath)
        PathName = WellPath.replace(InitialPath,"").split("_",1)
        Company = PathName[1].split("\\",1)
        DB_list_temp = [PathName[0], Company[0]]
        for Mnemonic in Mnemonic_head:
            try:
                DB_list_temp = DB_list_temp + [eval("Well.well." + Mnemonic + ".value")]
            except:
                DB_list_temp = DB_list_temp + [np.NaN]

        DB_list.append(DB_list_temp)
        print(PathName + ["-- Done"])

    WellHead_DB = pd.DataFrame(data=np.array(DB_list), columns = columns_name)

    return WellHead_DB

def reformatV2forPetrel(WellPathList, InitialPath, OutPath):
    """
    Match the Well name in the Mnemonic header with the filename.
    """
    for WellPath in WellPathList :
        Well = lasio.read(WellPath)
        PathName = WellPath.replace(InitialPath,"").split("_",1)[0]
        Well.well.WELL = PathName
        Well.write(OutPath + "\\" + PathName +".las", version=2)
        print("reformat " + PathName + " from " + WellPath + " to " + OutPath + "\\" + PathName +".las")
    pass

def reformatV2forWelly(WellPathList, OutPath):
    """
    Match the Well name in the Mnemonic header with the filename.
    """
    log = []
    for WellPath in WellPathList :
        Well = lasio.read(WellPath)
        # PathName = WellPath.replace(InitialPath,"").split(".la",1)[0]
        PathName = WellPath.split("\\")[-3]
        Well.well.UWI = PathName
        Well.write(OutPath + "\\" + PathName +".las", version=2)
        log_temp = "reformat " + PathName + " from " + WellPath + " to " + OutPath + "\\" + PathName +".las"
        log.append(log_temp)
        print(log_temp)
    return log
def Excel2Dict(Excel_DB, Keys_name='Keys', Value_name='Alias'):
    Dict_out = {}
    for Keys in Excel_DB[Keys_name].unique():
        Dict_out[Keys] = Excel_DB.loc[Excel_DB[Keys_name] == Keys, Value_name].to_list()
    return Dict_out
```
