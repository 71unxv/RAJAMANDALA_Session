```python
import pandas as pd
import numpy as np
import os
import glob
from IPython.display import display
```

```python
def getIndexPath(initialdir):
    """ 
    Search all the available Index.xls files and return its path for further directory references
    """    

    wellfolder = os.listdir(initialdir)
    largest_all = []
    for wellfolder_temp in wellfolder:
        path =initialdir +"\\" + wellfolder_temp 
#         print(path)
        if os.path.exists(path) and os.path.exists(path + "\\Index.xls"):
            if (len(os.listdir(path)) !=0):
                largest_all.append(sorted( (os.path.getsize(s), s) for s in glob.glob(path + "\\*.xls") )[-1][1])
    print("Total Index.xls : " + str(len(wellfolder)))
    print("Total Available Index.xls data : " + str(len(largest_all)))
    return largest_all

def BatchCollectIndex(AllIndexPath, OutPath):
    OutPathList = []
    for IndexPath in AllIndexPath:
        print(IndexPath + " : DONE                                                                                                            ", end="\r")
        Index = pd.read_excel(IndexPath)
        # WellPath.split("\\")[-1]
        Filename = IndexPath.split("\\")[-2] + "_Index"
        
        Index.to_csv(OutPath + Filename + ".csv")
        OutPathList.append(OutPath + Filename + ".csv")
                                        
    return OutPathList
```

```python
def getUniqueIndexWord(AllIndexPath):
    AllIndexWord = []
    for IndexPath in AllIndexPath:
#     IndexPath = AllIndexPath[1234]
        Index = pd.read_csv(IndexPath)
#         print(Index.iloc[:,6])
        try:
            AllIndexWord = AllIndexWord + list(Index.iloc[:,6])
        except:
            print("ERROR in " + IndexPath)
    return AllIndexWord


AllIndexWord = getUniqueIndexWord(AllIndexPath)
```

```python
import re
import string
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

AllResult = []
for i in AllIndexWord:
    if i is not np.nan: 
#         print(i)
        i = i.upper()
        result = re.sub(r'\d+', '', i)
        Tokenizer = RegexpTokenizer(r'\w+')
        result = result.replace("_", "")
    #     stop_words = set(stopwords.word('english'))
        result = Tokenizer.tokenize(result)              
        AllResult = AllResult + (result)

#     print((result))
AllResult = set([word for word in AllResult if word not in stopwords.words('english')])
AllResult = pd.DataFrame(AllResult)
AllResult.to_csv('RAJAMANDALA_UniqueIndexString.csv')
```

```python
def BatchScreenIndex(AllIndexPath, InitialPath, IndexKeys):
    OutPathList = []
    WellnameList = []
    XRDList= []
    XRDDescr= []
    SWCList= []
    SWCDescr= []
    PetroGraphList= []
    PetroGraphDescr= []
    MasterLogList= []
    MasterLogDescr= []
    GeoCompList= []
    GeoCompDescr= []
    GeoChemList= []
    GeoChemDescr= []
    DipMeterList= []
    DipMeterDescr= []
    CheckShotList= []
    CheckShotDescr= []
    BioStratList= []
    BioStratDescr= []
#     DirectoryList = []
    for IndexPath in AllIndexPath:
        

#         print(IndexPath)
        Index = pd.read_csv(IndexPath)
        Wellname = IndexPath.replace(InitialPath + "\\","").split("_Index.csv",1)[0] 
        WellnameList.append(Wellname)
        ###########3
        XRDList.append(' ')
        XRDDescr.append(' ')
        SWCList.append(' ')
        SWCDescr.append(' ')
        PetroGraphList.append(' ')
        PetroGraphDescr.append(' ')
        MasterLogList.append(' ')
        MasterLogDescr.append(' ')
        GeoCompList.append(' ')
        GeoCompDescr.append(' ')
        GeoChemList.append(' ')
        GeoChemDescr.append(' ')
        DipMeterList.append(' ')
        DipMeterDescr.append(' ')
        CheckShotList.append(' ')
        CheckShotDescr.append(' ')
        BioStratList.append(' ')
        BioStratDescr.append(' ')
        
        
#         print(Wellname)
        for Descr in Index.iloc[:,6]:
            Descr = str(Descr)
#             print(Wellname + "--" + str(type(Descr)) +"                                                                                                            ", end="\r")
            if any(ele in Descr for ele in IndexKeys["BIOSTRATIGRAPHY"]):	
                BioStratList[-1] = 'Available'
                BioStratDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["CHECKSHOT"]):	
                CheckShotList[-1] = 'Available'
                CheckShotDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["DIPMETER"]):	
                DipMeterList[-1] = 'Available'
                DipMeterDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["GEOCHEMICAL"]):	
                GeoChemList[-1] = 'Available'
                GeoChemDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["GEOLOGICAL COMPLETION"]):	
                GeoCompList[-1] = 'Available'
                GeoCompDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["MASTER LOG"]):	
                MasterLogList[-1] = 'Available'
                MasterLogDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["PETROGRAPHY"]):	
                PetroGraphList[-1] = 'Available'
                PetroGraphDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["SWC"]):	
                SWCList[-1] = 'Available'
                SWCDescr[-1] = Descr
            if any(ele in Descr for ele in IndexKeys["XRD"]):	
                XRDList[-1] = 'Available'
                XRDDescr[-1] = Descr

    IndexCheckList = pd.DataFrame(columns = ["WELLNAME",
                                             "BIOSTRATIGRAPHY",
                                                "BIOSTRATIGRAPHY DESCRIPTION",
                                                "CHECKSHOT",
                                                "CHECKSHOT DESCRIPTION",
                                                "DIPMETER",
                                                "DIPMETER DESCRIPTION",
                                                "GEOCHEMICAL",
                                                "GEOCHEMICAL DESCRIPTION",
                                                "GEOLOGICAL COMPLETION",
                                                "GEOLOGICAL COMPLETION DESCRIPTION",
                                                "MASTER LOG",
                                                "MASTER LOG DESCRIPTION",
                                                "PETROGRAPHY",
                                                "PETROGRAPHY DESCRIPTION",
                                                "SWC",
                                                "SWC DESCRIPTION",
                                                "XRD",
                                                "XRD DESCRIPTION"])
    IndexCheckList["WELLNAME"] = WellnameList
    IndexCheckList["BIOSTRATIGRAPHY"]= BioStratList
    IndexCheckList["BIOSTRATIGRAPHY DESCRIPTION"]= BioStratDescr
    IndexCheckList["CHECKSHOT"]= CheckShotList
    IndexCheckList["CHECKSHOT DESCRIPTION"]= CheckShotDescr
    IndexCheckList["DIPMETER"]= DipMeterList
    IndexCheckList["DIPMETER DESCRIPTION"]= DipMeterDescr
    IndexCheckList["GEOCHEMICAL"]= GeoChemList
    IndexCheckList["GEOCHEMICAL DESCRIPTION"]= GeoChemDescr
    IndexCheckList["GEOLOGICAL COMPLETION"]= GeoCompList
    IndexCheckList["GEOLOGICAL COMPLETION DESCRIPTION"]= GeoCompDescr
    IndexCheckList["MASTER LOG"]= MasterLogList
    IndexCheckList["MASTER LOG DESCRIPTION"]= MasterLogDescr
    IndexCheckList["PETROGRAPHY"]= PetroGraphList
    IndexCheckList["PETROGRAPHY DESCRIPTION"]= PetroGraphDescr
    IndexCheckList["SWC"]= SWCList
    IndexCheckList["SWC DESCRIPTION"]= SWCDescr
    IndexCheckList["XRD"]= XRDList
    IndexCheckList["XRD DESCRIPTION"]= XRDDescr
    return IndexCheckList

# Path ="D:\\EKSINDO\\EKSINDO_CompiledIndex"
CompiledIndexPath = "RAJAMANDALA_CompiledIndex"
CompiledIndexPath_list = glob.glob(CompiledIndexPath + "\\*.csv")

IndexCheckList = BatchScreenIndex(CompiledIndexPath_list, CompiledIndexPath, IndexKeys)
IndexCheckList.to_csv('RAJAMANDALA_IndexCheckList_All.csv')
```
